```python
import random

def generar_posiciones_vehiculos(coordenadas):
    vehiculos = []
    # Iterar por la lista de coordenadas de los vehículos
    for i, coord_vehiculo in enumerate(coordenadas):
        mensaje_vehiculo = {"vehiculo"+str(i): coord_vehiculo}
        vehiculos.append(mensaje_vehiculo)
    return vehiculos

lista_mensajes_vehiculos = generar_posiciones_vehiculos(data)

for mensaje in lista_mensajes_vehiculos:
    print(mensaje)

def generar_posiciones_solicitantes(coordenadas):
    solicitudes = []
    # Mezclar las coordenadas de los solicitantes aleatoriamente
    coord_solicitante = coordenadas.copy()  # Hacemos una copia de los datos de los vehículos
    random.shuffle(coord_solicitante)
    for i, coord_solicitante in enumerate(coordenadas):
        mensaje_solicitante = {"Solicitante"+str(i): coord_solicitante}
        solicitudes.append(mensaje_solicitante)
    return solicitudes


lista_mensajes_solicitudes = generar_posiciones_solicitantes(data)



for mensaje2 in lista_mensajes_solicitudes:
    print(mensaje2)


import apache_beam as beam
from apache_beam.options.pipeline_options import PipelineOptions
from apache_beam.options.pipeline_options import PipelineOptions

pipeline_options = PipelineOptions(['--direct_running_mode=direct'])


with beam.Pipeline(options=PipelineOptions()) as pipeline:
        mensajes_vehiculos = (
            pipeline
            | "Generar Mensajes Vehículos" >> beam.Create(lista_mensajes_vehiculos)
            |"Imprimir mensajes" >> beam.Map(print)
        )
```
